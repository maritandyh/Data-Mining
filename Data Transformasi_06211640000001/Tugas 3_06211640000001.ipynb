### Pre-processing Data
### a. Missing Value

```python
import pandas as pd
import numpy as np
```

```python
datacsm = pd.read_csv("E:\Kuliah\Semester 6\DATMIN\Data Transformasi_06211640000001\CSM dataset 2014 and 2015.csv")
np.sum(datacsm.isnull())
```

```python
datacsmdes=datacsm.drop(['Movie','Year'],1)
datacsmdes.describe()
```

```python
datacsm['Budget']=datacsm['Budget'].fillna(datacsm['Budget'].mean())
datacsm['Screens']=datacsm['Screens'].fillna(datacsm['Screens'].mean())
datacsm['Aggregate Followers']=datacsm['Aggregate Followers'].fillna(datacsm['Aggregate Followers'].mean())
np.sum(datacsm.isnull())
```

```python
datacsmnomiss = datacsm
datacsmnomiss.head()
```

```python
datacsmnomiss.to_excel('CSM dataset 2014 and 2015 tanpa missing value.xlsx')
```

### b. Missing Value, Outlier

```python
from scipy import stats
data1=datacsm.drop(['Movie','Year'],1)
z = np.abs(stats.zscore(data1._get_numeric_data()))
print(z)
datacsmnomissout=datacsm[(z<3).all(axis=1)]
print(datacsmnomissout.shape)
```

```python
datacsmnomissout.head(10)
```

```python
datacsmnomissout.to_excel('CSM dataset 2014 and 2015 tanpa missing value, outlier.xlsx')
```

### c.Missing Value, Transformasi

```python
x = datacsm.iloc[0:217,2:11]
x.head()
```

```python
#normalization
from sklearn.preprocessing import normalize
xnorm = normalize(x,norm='l2') #default norm L2 and max
xnorm = pd.DataFrame(xnorm)
xnorm.tail()
```

```python
# Softmax
## Softmax merupakan metode normalisasi pengembangan transformasi secara linier. 
## Metode ini sangat berguna pada saat data-data yang ada melibatkan data-data outlier.
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x), axis = 0)
xsoft = softmax(xnorm)
xsoft = pd.DataFrame(xsoft)
print(xsoft)
```

```python
xsoft.to_excel('CSM dataset 2014 and 2015 tanpa missing value dan telah ditransformasikan.xlsx')
```

### d. Missing Value, Outlier, Transformasi

```python
xx = datacsmnomissout.iloc[0:198,2:11]
xx.head()
```

```python
#normalization
from sklearn.preprocessing import normalize
xxnorm = normalize(xx,norm='l2') #default norm L2 and max
xxnorm = pd.DataFrame(xxnorm)
xxnorm.tail()
```

```python
# Softmax
def softmax(xx):
    return np.exp(xx)/np.sum(np.exp(xx), axis = 0)
xxsoft = softmax(xxnorm)
xxsoft = pd.DataFrame(xxsoft)
print(xxsoft)
```
