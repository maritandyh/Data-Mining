```python
import pandas as pd
import numpy as np
```

```python
datasydney = pd.read_csv("E:\Kuliah\Semester 6\DATMIN\Minggu ke-6\weatherAUS.csv")
np.sum(datasydney.isnull())
```

```python
datasydneydes=datasydney.drop(['Date','Location'],1)
datasydneydes.describe()
```

```python
datasydney['MinTemp']=datasydney['MinTemp'].fillna(datasydney['MinTemp'].mean())
datasydney['MaxTemp']=datasydney['MaxTemp'].fillna(datasydney['MaxTemp'].mean())
datasydney['Rainfall']=datasydney['Rainfall'].fillna(datasydney['Rainfall'].median())
datasydney['Evaporation']=datasydney['Evaporation'].fillna(datasydney['Evaporation'].mean())
datasydney['Sunshine']=datasydney['Sunshine'].fillna(datasydney['Sunshine'].mean())
datasydney['WindSpeed9am']=datasydney['WindSpeed9am'].fillna(datasydney['WindSpeed9am'].mean())
datasydney['WindSpeed3pm']=datasydney['WindSpeed3pm'].fillna(datasydney['WindSpeed3pm'].mean())
datasydney['Humidity9am']=datasydney['Humidity9am'].fillna(datasydney['Humidity9am'].mean())
datasydney['Humidity3pm']=datasydney['Humidity3pm'].fillna(datasydney['Humidity3pm'].mean())
datasydney['Pressure9am']=datasydney['Pressure9am'].fillna(datasydney['Pressure9am'].mean())
datasydney['Pressure3pm']=datasydney['Pressure3pm'].fillna(datasydney['Pressure3pm'].mean())
datasydney['Cloud9am']=datasydney['Cloud9am'].fillna(datasydney['Cloud9am'].mean())
datasydney['Cloud3pm']=datasydney['Cloud3pm'].fillna(datasydney['Cloud3pm'].mean())
datasydney['Temp9am']=datasydney['Temp9am'].fillna(datasydney['Temp9am'].mean())
datasydney['Temp3pm']=datasydney['Temp3pm'].fillna(datasydney['Temp3pm'].mean())
datasydney['RISK_MM']=datasydney['RISK_MM'].fillna(datasydney['RISK_MM'].median())
datasydney['WindGustSpeed']=datasydney['WindGustSpeed'].fillna(datasydney['WindGustSpeed'].mean())
np.sum(datasydney.isnull())
```

```python
from scipy.stats import mode
datasydney['WindGustDir']=datasydney['WindGustDir'].fillna(datasydney['WindGustDir']).dropna().mode()[0]
datasydney['WindDir9am']=datasydney['WindDir9am'].fillna(datasydney['WindDir9am']).dropna().mode()[0]
datasydney['WindDir3pm']=datasydney['WindDir3pm'].fillna(datasydney['WindDir3pm']).dropna().mode()[0]
datasydney['RainToday']=datasydney['RainToday'].fillna(datasydney['RainToday']).dropna().mode()[0]
datasydney['RainTomorrow']=datasydney['RainTomorrow'].fillna(datasydney['RainTomorrow']).dropna().mode()[0]
np.sum(datasydney.isnull())
```

```python
from sklearn.preprocessing import normalize
xxnorm = normalize(xx,norm='l2') #default norm L2 and max
xxnorm = pd.DataFrame(xxnorm)
xxnorm.head()
```

```python
# Softmax
def softmax(xx):
    return np.exp(xx)/np.sum(np.exp(xx), axis = 0)
xxsoft = softmax(xxnorm)
xxsoft = pd.DataFrame(xxsoft)
print(xxsoft)
```
